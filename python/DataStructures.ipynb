```python
def reverse_ll(head):
    if not (head and head.next) :
        return head
    
    prev = None
    while head is not None:
        nextNode = head.next

        head.next = prev
        prev = head
        head = nextNode

    return prev
```

```python
def replace_max_val(head, val):
    if not head:
        return head
    ptr = head

    curr = head
    while curr is not None:
        if curr.data > ptr.data:
            ptr = curr
        curr = curr.next
    ptr.data = val
    return head
        
def sum_odd_nodes(head):
    if head is None or head.next.next is None:
        return 0
    total_sum = 0
    curr = head.next
    while (curr.next is not None and curr.next.next is not None):
        total_sum += curr.data
        curr = curr.next.next

    total_sum += curr.data
    return total_sum

def apply_rules(head):
    # if * and / appears int the Begning or at the end remove it
    # if * appears remove it with space 
    # if * & / appears consecutilvely remove it
    curr = head
    while curr is not None:
        if curr.data == '*' or curr.data == '/':
            curr.data = ' '
            if curr.next.data == '/' or curr.next.data == '*':
                curr.next = curr.next.next
                curr.next.data = curr.next.data.capitalize()
        curr = curr.next
```
