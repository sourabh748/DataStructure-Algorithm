```python
def ispalindrom(string, start, end):
    if end <= start:
        return True
    
    if string[start] == string[end]:
        return palindrom(string, start +1, end -1)
    
    return False
```

```python
def decimalToBinary(nums : int, result: str = ""):
    if nums == 0:
        return result

    result = str(nums % 2) + result
    return decimalToBinary(nums//2, result)
```

```python
def sumOfnNaturalNumber(nums: int):
    if nums <= 1:
        return nums
    return nums + sumOfnNaturalNumber(nums -1)
```

```python
def binarySearch(arr: list, target: int, left: int, right: int):
    if left > right:
        return -1
    mid = right - (right - left)//2

    if arr[mid] == target:
        return mid

    if arr[mid] < target:
        return binarySearch(arr, target, mid +1, right)
    return binarySearch(arr, target, left, mid -1)
    
```

```python
def mergeSort(arr: list):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    
    left = mergeSort(arr[: mid])
    right = mergeSort(arr[mid :])

    leftIndex = rightIndex = 0
    arr = []

    while leftIndex < len(left) and rightIndex < len(right):
        if left[leftIndex] >= right[rightIndex]:
            arr.append(right[rightIndex])
            rightIndex += 1
            continue 
        arr.append(left[leftIndex])
        leftIndex += 1

    while leftIndex < len(left):
        arr.append(left[leftIndex])
        leftIndex += 1
    
    while rightIndex < len(right):
        arr.append(right[rightIndex])
        rightIndex += 1

    return arr
```
